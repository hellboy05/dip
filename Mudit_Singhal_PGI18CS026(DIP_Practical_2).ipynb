```python
k1 = np.array([[1, 0, -1],
               [0, 0, 0],
               [-1, 0, 1]])
k2 = np.array([[-1, 0, -1],
               [0, 4, 0],
               [-1, 0, -1]])
k3 = np.array([[-1, -1, -1],
               [-1, 8, -1],
               [-1, -1, -1]])
```

```python
filter2 = cv2.filter2D(img, -1, k1)
filter3 = cv2.filter2D(img, -1, k2)
filter4 = cv2.filter2D(img, -1, k3)
```
